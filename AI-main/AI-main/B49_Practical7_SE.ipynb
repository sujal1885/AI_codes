```python
data = pd.read_csv('play_tennis.csv')
```

```python
class_counts = data['play'].value_counts()
class_total = class_counts.sum()

features = ['outlook', 'temp', 'humidity', 'wind']
for feature in features:
    feature_values = data[feature].unique()
    for value in feature_values:
        for class_label in class_counts.index:
            subset = data[data['play'] == class_label]
            count = subset[subset[feature] == value].shape[0]
            total = class_counts[class_label]
            probability = count / total
            print(f"P({value}|{class_label}) = {count}/{total} = {probability}")
```

```python
def naive_bayes_probability(data, feature_values, alpha=1):

    class_counts = data['play'].value_counts(normalize=True)
    play_yes_prob = class_counts['Yes']
    play_no_prob = class_counts['No']


    conditional_prob_yes = 1
    conditional_prob_no = 1


    for i, value in enumerate(feature_values):

        subset_yes = data[(data['play'] == 'Yes') & (data.iloc[:, i + 1] == value)]
        subset_no = data[(data['play'] == 'No') & (data.iloc[:, i + 1] == value)]


        conditional_prob_yes *= (len(subset_yes) + alpha) / (len(data[data['play'] == 'Yes']) + alpha * len(feature_values))
        conditional_prob_no *= (len(subset_no) + alpha) / (len(data[data['play'] == 'No']) + alpha * len(feature_values))


    naive_bayes_prob_yes = conditional_prob_yes * play_yes_prob
    naive_bayes_prob_no = conditional_prob_no * play_no_prob

    return naive_bayes_prob_yes, naive_bayes_prob_no

data = pd.read_csv('play_tennis.csv')

feature_values = ['sunny', 'hot', 'normal', False]


naive_bayes_prob_yes, naive_bayes_prob_no = naive_bayes_probability(data, feature_values)

print("Naive Bayes Probability for 'Yes' (Play Tennis):", naive_bayes_prob_yes)
print("Naive Bayes Probability for 'No' (Don't Play Tennis):", naive_bayes_prob_no)
```
