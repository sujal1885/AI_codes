```python
#8 puzzlepuzzle
import time
from collections import deque

def count_parity(puzzle):
    parity = {}
    k = 0
    for i in range(len(puzzle)):
        element = puzzle[i]
        count = 0
        for j in range(i + 1, len(puzzle)):
            next_element = puzzle[j]
            if next_element == 0:
                continue
            elif element > next_element:
                count += 1
        parity[element] = count
    print(parity)

def is_solvable(total_parity):
    if (total_parity % 2 == 0):
        print("Solvable")
    else:
        print("Unsolvable")

def print_puzzle(board):
    for row in board:
        print(" ".join(map(str, row)))
    print()

def find_blank(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return i, j

def is_goal(board):
    return board == [[1, 2, 3], [4, 5, 6], [7, 8, 0]]

def move_left(board):
    i, j = find_blank(board)
    if j > 0:
        board[i][j], board[i][j - 1] = board[i][j - 1], board[i][j]
        return board
    return None

def move_right(board):
    i, j = find_blank(board)
    if j < 2:
        board[i][j], board[i][j + 1] = board[i][j + 1], board[i][j]
        return board
    return None

def move_up(board):
    i, j = find_blank(board)
    if i > 0:
        board[i][j], board[i - 1][j] = board[i - 1][j], board[i][j]
        return board
    return None

def move_down(board):
    i, j = find_blank(board)
    if i < 2:
        board[i][j], board[i + 1][j] = board[i + 1][j], board[i][j]
        return board
    return None

def bfs(start):
    frontier = deque([(start, [], 0)])  # Modification to keep track of path
    explored = set()

    while frontier:
        current_state, path, steps = frontier.popleft()

        explored.add(tuple(map(tuple, current_state)))
        for move_func, move_name in [(move_left, "left"), (move_right, "right"), (move_up, "up"), (move_down, "down")]:
            new_state = move_func([row[:] for row in current_state])
            if new_state and tuple(map(tuple, new_state)) not in explored:
                new_path = path + [move_name]  # Update path with the current move
                if is_goal(new_state):
                    return new_state, new_path, steps + 1
                frontier.append((new_state, new_path, steps + 1))
            # print(new_state)

    return None, [], 0

initial_state = [
    [1, 2, 3],
    [0, 4, 5],
    [6, 7, 8]
]

parity = {}

flat_list = sum(initial_state, [])

print(flat_list)

count_parity(flat_list)

total_parity = 0
for values in parity.values():
    total_parity = total_parity + values

print(f"Total parity is {total_parity}")

is_solvable(total_parity)
start = time.time()
solution, path, steps = bfs(initial_state)  # Unpack three values from bfs function
end = time.time()

total_time = end - start
if solution:
    print("Solution found in", steps, "steps:")
    print(f"Total Time required is {total_time}")
    print_puzzle(solution)
    print("Path taken:")
    print(path)
else:
    print("No solution found.")
```
