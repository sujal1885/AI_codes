```python
initial = [[1,2,3],[4,0,6],[7,5,8]]
final = [[1,2,3],[4,5,6],[7,8,0]]
rows = 3
cols = 3

def move_up(G):
    for i in range(rows):
        if G[0][i] == 0:
            return None

    for i in range(rows):
        for j in range(cols):
            if G[i][j] == 0:
                G[i][j], G[i - 1][j] = G[i - 1][j], 0
                return G

def move_down(G):
    for i in range(rows):
        if G[rows-1][i] == 0:
            return None

    for i in range(rows):
        for j in range(cols):
            if G[i][j] == 0:
                G[i][j], G[i + 1][j] = G[i + 1][j], 0
                return G

def move_right(G):
    for i in range(rows):
        if G[i][cols-1] == 0:
            return None

    for i in range(rows):
        for j in range(cols):
            if G[i][j] == 0:
                G[i][j], G[i][j + 1] = G[i][j + 1], 0
                return G

def move_left(G):
    for i in range(rows):
        if G[i][0] == 0:
            return None

    for i in range(rows):
        for j in range(cols):
            if G[i][j] == 0:
                G[i][j], G[i][j - 1] = G[i][j - 1], 0
                return G

def goaltest(G):
    for i in range(rows):
        for j in range(cols):
            if G[i][j] != final[i][j]:
                return False
    return True

def BFS(initial, final):
    explored = set()
    frontier = [initial]
    states_generated = 0
    level = 1
    curr = 1

    while frontier:
        current_state = frontier.pop(0)
        explored.add(tuple(tuple(row) for row in current_state))

        print("Current State:")
        for row in current_state:
            print(row)
        print()

        states_generated += 1

        if goaltest(current_state):
            print("Goal State Reached!")
            print("States Generated:", states_generated)
            return True

        for move_func in [move_right, move_left, move_up, move_down]:
            new_state = [list(row) for row in current_state]
            result = move_func(new_state)
            if result and tuple(map(tuple, result)) not in explored and result not in frontier:
                frontier.append(tuple(map(tuple, result)))
                explored.add(tuple(map(tuple, result)))
        curr -= 1
        if curr == 0:
          print('Level ',level)
          level+=1
          curr = len(frontier)
          print('Number of states in level ',level,' is ',curr)

    print("No solution found.")
    return False

result = BFS(initial, final)
print("Solution found:", result)
```
